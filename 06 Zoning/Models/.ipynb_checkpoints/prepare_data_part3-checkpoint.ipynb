```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_excel("../Data/zoning_final.xlsx")
```

```python
df.head(3)
```

```python
df["Review - Main_Sentence"][0]
```

```python
import nltk.data
nltk.download('punkt')
```

```python
list_master_sentence = []
for i in range(0,len(df["Review - Main_Sentence"])):
    text = df["Review - Main_Sentence"][i]
    a_list = nltk.tokenize.sent_tokenize(text)
    list_master_sentence.append(a_list)
```

```python
df["Sentence - Main_Sentence"] = list_master_sentence
```

```python
df.head(3)
```

```python
list_full_sentence = [item for sublist in list_master_sentence for item in sublist]
```

```python
len(list_full_sentence)
```

```python
df_repeated = pd.concat([df]*7, ignore_index=True)
display(df_repeated.head(3))
display(df_repeated.shape)
```

```python
list_daframe = []
for i in range(0,len(df["Sentence - Main_Sentence"])):
    list_daframe.append(df_final.append([df.loc[i:i]]*len(df["Sentence - Main_Sentence"][i]),ignore_index=True))
```

```python
df_final_v2 = pd.concat(list_daframe)
display(df_final_v2.head(3))
display(df_final_v2.shape)
```

```python
df_final_v2["Sentence - Main_Sentence"] = list_full_sentence
```
