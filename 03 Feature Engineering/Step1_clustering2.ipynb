```python
list_counter=[]
for i in range(0,len(df1["Review"])):
    c = 0
    for j in list_spectator:
        if type(df1["Review"][i]) != float:
            if j in df1["Review"][i].lower():
                c= c+1
    list_counter.append(c)
df1["Spectator_Cat"] = list_counter
```
