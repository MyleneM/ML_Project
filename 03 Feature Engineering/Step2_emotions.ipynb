# Step2_emotions

--------------------------------------
The goal of this script is to perform sentiment and keywords analysis on the Theatre's Reviews

## I) Load Libraries

```python
import pandas as pd
import numpy as np
from textblob import TextBlob
import text2emotion as te
import nltk
from gensim.summarization import keywords
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

## II) Load Data

```python
df = pd.read_excel("corpus_I_2clustering.xlsx")
display(df.head(3))
display(df.shape)
```

## III) Sentiment Analysis

```python
sid = SentimentIntensityAnalyzer()
```

```python
df2 = df.copy()
list_neg = []
list_neu = []
list_pos = []
list_compound = []
list_polarity = []
list_subjectivity = []
for i in range(0,len(df2["Review"])):
    score1 = sid.polarity_scores(df2["Review"][i])
    score2 = TextBlob(df2["Review"][i]).sentiment
    list_neg.append(score1['neg'])
    list_neu.append(score1['neu'])
    list_pos.append(score1['pos'])
    list_compound.append(score1['compound'])
    list_polarity.append(score2[0])
    list_subjectivity.append(score2[1])
```

```python
df2["Neg"] = list_neg
df2["Neu"] = list_neu
df2["Pos"] = list_pos
df2["Compound"] = list_compound
df2["Polarity"] = list_polarity
df2["Subjectivity"] = list_subjectivity
```

```python
display(df2.head(3))
display(df2.shape)
```

```python
df_corpus_v3 = df2
df_corpus_v3["Emotion"] = ""
df_corpus_v3["Emotion"][0:5] = df_corpus_v3["Review"][0:5].apply(lambda x:te.get_emotion(x))
```

```python
list_emotion_angry = []
list_emotion_fear = []
list_emotion_happy = []
list_emotion_sad = []
list_emotion_surprise = []
for i in range(0,len(df_corpus_v3["Review"])):
  print(i)
  emotion =te.get_emotion(df_corpus_v3["Review"][i])
  list_emotion_angry.append(emotion["Angry"])
  list_emotion_fear.append(emotion["Fear"])
  list_emotion_happy.append(emotion["Happy"])
  list_emotion_sad.append(emotion["Sad"])
  list_emotion_surprise.append(emotion["Surprise"])
```

```python
df_corpus_v3["Angry"] = list_emotion_angry
df_corpus_v3["Fear"] = list_emotion_fear
df_corpus_v3["Happy"] = list_emotion_happy
df_corpus_v3["Sad"] = list_emotion_sad
df_corpus_v3["Surpise"] = list_emotion_surprise
```

```python
df_corpus_v3.to_excel("corpus_I_2clustering_emotions.xlsx")
```

## IV) Keywords Analysis

```python
df_corpus_v4 = df_corpus_v3
```

```python
list_keywords = []
for i in range(0,len(df_corpus_v4['Review'])):
    print(i)
    text_en = (df_corpus_v4['Review'][i])
    list_keywords.append(keywords(text_en,words = 10,scores = True, lemmatize = True))
```

```python
df_corpus_v4['Top_10_Keywords'] = list_keywords
```
