## III) Statistical Analysis

```python
df_II["Start_Review_1"] = ""
df_II["End_Review_1"] = ""
df_II["Start_Review_2"] = ""
df_II["End_Review_2"] = ""
df_II["Start_Review_3"] = ""
df_II["End_Review_3"] = ""
df_II["Nbr_of_Sentences"] = ""
```

```python
start_review_1 = []
start_review_2 = []
start_review_3 = []
end_review_1 = []
end_review_2 = []
end_review_3 = []
nbr_of_sentences = []
for i in range(0,len(df_II["Review"])):
    if len(df_II["Review"][i].split(".")) > 6:
        nbr_of_sentences.append(len(df_II["Review"][i].split(".")))
        start_review_1.append(df_II["Review"][i].split(".")[0:1])
        start_review_2.append(df_II["Review"][i].split(".")[0:2])
        start_review_3.append(df_II["Review"][i].split(".")[0:3])
        end_review_1.append(df_II["Review"][i].split(".")[-2:-1])
        end_review_2.append(df_II["Review"][i].split(".")[-3:-1])
        end_review_3.append(df_II["Review"][i].split(".")[-4:-1])
        
    else:
        nbr_of_sentences.append(len(df_II["Review"][i].split(".")))
        start_review_1.append("n.a")
        start_review_2.append("n.a")
        start_review_3.append("n.a")
        end_review_1.append("n.a")
        end_review_2.append("n.a")
        end_review_3.append("n.a")
```

```python
df_II.shape
```

```python
df_II["Start_Review_1"] = start_review_1
df_II["End_Review_1"] = end_review_1
df_II["Start_Review_2"] = start_review_2
df_II["End_Review_2"] = end_review_2
df_II["Start_Review_3"] = start_review_3
df_II["End_Review_3"] = end_review_3
df_II["Nbr_of_Sentences"] = nbr_of_sentences
df_II.head(3)
```

```python
word_average_start_1 = []
for i in range(0,len(df_II_v2["Start_Review_1"])):
    if df_II_v2["Start_Review_1"][i][0] != '':
        words = df_II_v2["Start_Review_1"][i][0].split()
        if len(words) > 0:
            average = sum(len(word) for word in words) / len(words)
            avg = round(average,2)
            word_average_start_1.append(avg)
        else:
            word_average_start_1.append('n.a')
    else:
        word_average_start_1.append('n.a')
```
